```python
from bs4 import BeautifulSoup as soup
import requests
```

```python
from urllib.request import urlopen as ureq
```

```python
my_url='https://www.newegg.com/Video-Cards-Video-Devices/Category/ID-38?Tpk=video%20cards'
```

```python
my_url
```

```python
#opening up connection, grabbing the page
```

```python
u_client=ureq(my_url)
```

```python
pagehtml=u_client.read() #read the page
```

```python
u_client.close() #close the connection
```

```python
page_soup= soup(pagehtml,'html.parser') # this does the html parsing
```
